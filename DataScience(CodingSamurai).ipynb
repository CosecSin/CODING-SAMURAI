# **Cleaning The Listing Data**

---



```python
# Total Number of Null Values of each columns
listing_df.isnull().sum()
```

```python
#Calculating The Percentage Of Missing Values
listings_df_miss = pd.DataFrame((listing_df.isnull().sum())*100/len(listing_df), columns=[ '% Missing Values'])
listings_df_miss[listings_df_miss['% Missing Values']>0]
```

```python
# As ["square_feet" , "license"] columns contain mostly null values we will drop them.
listing_df = listing_df.drop(columns = ["square_feet" , "license"])
```

```python
#Remove redundent columns
listing_df = listing_df.drop('scrape_id', axis=1)
listing_df.info()
```

```python
missing_cols = ['security_deposit','cleaning_fee','review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value','reviews_per_month','monthly_price','price']
listing_df[missing_cols].info()
```

```python
# Fill nulls with zeros
missing_price = ['security_deposit','weekly_price','cleaning_fee','monthly_price','price']
for i in missing_price:
    listing_df[i] = listing_df[i].str.replace('$','')
    listing_df[i] = listing_df[i].str.replace(',','')
    listing_df[i] = listing_df[i].astype('float')
```

```python
# Replacing weekly_price Null Values With price*6.
listing_df['weekly_price'].fillna(listing_df['price'] * 6, inplace=True)
```

```python
# Replacing monthly_price Null Values with weekly_price *3.
listing_df['monthly_price'] = listing_df['monthly_price'].fillna(listing_df['weekly_price']*3, inplace= False)
```

```python
clean_price = listing_df[['price','weekly_price','monthly_price']]
clean_price
```

```python
# Filling the Null Values of Security Deposit with mean values.
listing_df['security_deposit'] = listing_df['security_deposit'].fillna(listing_df['security_deposit'].mean(), inplace=False)
```

```python
# Filling the Null Values of cleaning_fee with mean values.
listing_df['cleaning_fee'] = listing_df['cleaning_fee'].fillna(listing_df['cleaning_fee'].mean(), inplace=False)
```

```python
listing_df[missing_price].head()
```

```python
# Select numeric columns for correlation analysis
numeric_columns = listing_df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()
```

# **Reading The DataSet Of Calender**

---



```python
calendar_df = pd.read_csv("calendar.csv", parse_dates = ['date'])
# Exploring The DataSet
calendar_df
```

```python
calendar_df.info()
```

**Data Cleaning**

```python
# First We Have To Remove '$' Symbol from the price.
calendar_df['price'].replace('\$|,','',regex = True, inplace = True)
```

```python
calendar_df
```

# **Cleaning The Calendar Data**

---



```python
#Converting The Price as Float To Calculate Mean
calendar_df['price'] = calendar_df['price'].astype(float)
```

```python
# Filling Null Values With Mean Price
calendar_df['price'].fillna(calendar_df['price'].mean(), inplace=True)
```

```python
calendar_df
```

```python
# Replacing All the T and F with 1 and 0 in the Available Column.
calendar_df['available'] = calendar_df['available'].replace({'t':1,'f':0})
```

```python
calendar_df['month'] = calendar_df['date'].dt.month
```

```python
# Changing The Munber of Month Into their names.
calendar_df['month'] = calendar_df['month'].replace({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May' ,6:"June" , 7:"July" , 8: "August" , 9:"September" , 10: "October" , 11:"November" , 12:"December"})
```

```python
calendar_df
```

# **Visualization And Statistics Of Calender**

---


```python
# Count the number of available and unavailable listings.
available_count = calendar_df['available'].value_counts()[1]
unavailable_count = calendar_df['available'].value_counts()[0]

# Create a pie chart
labels = ['Available', 'Unavailable']
sizes = [available_count, unavailable_count]
colors = ['skyblue', 'lightgreen']
```

```python
plt.pie(sizes,labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Available vs. Unavailable Listings ')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.countplot(x = 'month',hue = 'available',data=calendar_df, palette='bright')
plt.title('Available Rooms For Each Month')
plt.xlabel('Months')
plt.ylabel('Total Number of Rooms')
plt.xticks(rotation=45)
plt.legend(title='Availability')
plt.show()
```

```python
# Storing Only The Available Rooms
avail = calendar_df['available'] == 1
```

```python
# Creating A Data Frame For Avialable room
avail_df = calendar_df.loc[avail,:]
```

```python
avail_df
```

```python
plt.figure(figsize=(12, 8))
ax_1 = sns.boxplot(data = avail_df, x = 'month', y ='price', notch = True)
ax_1.set_xticklabels(ax_1.get_xticklabels())
ax_1.set_ylim(0,400)
plt.xticks(rotation=0)
plt.show()

```

```python
plt.figure(figsize=(12, 8))
listing_group = avail_df.groupby('month').count()
ax_2 = sns.barplot(x= listing_group.index.values,y =listing_group['listing_id'] )
ax_2.set_ylabel("Number of listing")
ax_2.set_xlabel("Month")
ax_2.set_title("Distribution of listing Monthly")
plt.xticks(rotation=40)
plt.show()
```

```python
# Extract the sentiment of each comment
review_df['sentiment'] = review_df['comments'].apply(lambda comment: TextBlob(comment).sentiment.polarity)

# Filter positive and negative comments
positive_comments = review_df[review_df['sentiment'] > 0]
negative_comments = review_df[review_df['sentiment'] < 0]
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    [len(positive_comments), len(negative_comments)],
    labels=['Positive', 'Negative'],
    autopct='%1.1f%%',
    startangle=140,
)
plt.title('Pie Chart of Positive and Negative Comments')
plt.axis('equal')
plt.show()
```
